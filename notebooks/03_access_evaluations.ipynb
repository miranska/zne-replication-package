# Accessing Evaluation Archive for the Synthetic Dataset

This archive contains condensed evaluation results for zero-noise extrapolation (ZNE) experiments.
The 5,400 files are stored in Apache Parquet format and capture the mitigated expectation values produced by the extrapolation models evaluated in the paper.

## 1. Included Data

Each Parquet file contains information about ZNE-mitigated expectation values together with the metadata needed to identify the circuit, observable, model, and execution context.

### Column Descriptions

- **`case_id`**: Unique identifier for the evaluation instance.
- **`model`**: Internal name of the extrapolation model.
  - `bounded_polynomial_extrapolation`: bounded polynomial extrapolation model.
  - `poly_extrapolation_mitiq`: unbounded polynomial extrapolation model.
  - `bounded_exp_extrapolation_minimize`: bounded exponential extrapolation model.
  - `exponential_extrapolation_mitiq`: unbounded exponential extrapolation model.
  - `bounded_polyexp_extrapolation_minimize`: bounded polynomial--exponential extrapolation model.
  - `polyexp_extrapolation_mitiq`: unbounded polynomial--exponential extrapolation model.
- **`model_label`**: Model label with parameters.
- **`scale_factors`**: Noise scale factors used for the extrapolation fit.
- **`degree`**: Polynomial degree or analogous model-order parameter, where applicable.
- **`asymptote`**: Asymptotic parameter used by a model, where applicable.
- **`observable`**: Observable whose expectation value is being estimated.
- **`num_qubits`**: Number of qubits in the circuit.
- **`generator_circuit_depth`**: Target circuit depth used during circuit generation by Qiskit.
- **`actual_circuit_depth`**: Depth of the compiled or executable circuit used in evaluation.
- **`actual_gate_count`**: Gate count of the compiled or executable circuit used in evaluation.
- **`runtime_provider`**: Backend  associated with the evaluation.
- **`circuit_path`**: Path identifying the circuit associated with the evaluation record in `qasm_archive.zip`.
- **`status`**: Execution or fitting status for the evaluation record (`ok` or `failed`).
- **`failure_reason`**: Reason for failure, if the record did not complete successfully.
- **`ideal_expectation`**: Ideal expectation value for the observable.
- **`mitigated_expectation`**: Mitigated expectation value produced by the extrapolation model.

---

## 2. Interpretation

Each row represents the output of one extrapolation model applied to one evaluation instance. The key quantity of interest is:

- **`mitigated_expectation`**, the ZNE-corrected estimate produced by the model,

which can be compared against:

- **`ideal_expectation`**, the corresponding ideal expectation value.

Together, these fields support downstream analyses such as error computation, matched bounded/unbounded comparisons, and aggregate performance evaluation.



```python
import pyarrow.dataset as ds

# Lazily read the dataset
dataset = ds.dataset("../data/evaluation_archive", format="parquet")

# Scan only the first 10 rows
table = dataset.head(10)

# Show the first 10 rows in a Pandas dataframe
table.to_pandas()
```
