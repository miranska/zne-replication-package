# Accessing Synthetic QASM Code

This notebook explores the contents of the archive `../data/qasm_archive.zip`.

```python
from pathlib import Path

archive_path = Path("../data/qasm_archive.zip")

# specify the parameters needed to access the QASM listing
n_qubits = 9
depth = 50
bin_value = -0.05
circuit_id = 0

# Construct directory and filenames

directory = f"{n_qubits}qubits_0.05bin_width_100circuits_per_bin_{depth}circuit_depth"
raw_filename = f"circuit_{bin_value:.2f}_{circuit_id}_raw.qasm3.qasm"
transpiled_filename = f"circuit_{bin_value:.2f}_{circuit_id}.qasm3.qasm"
raw_path = f"{directory}/{raw_filename}"
transpiled_path = f"{directory}/{transpiled_filename}"

print("Raw file path:        ", raw_path)
print("Transpiled file path: ", transpiled_path)
```

```python
def print_head_from_zip(zip_file, path, n_lines=10):
    """Helper to print first N lines"""
    print("\n" + "=" * 100)
    print(path)
    print("=" * 100)
    try:
        content = zip_file.read(path).decode("utf-8", errors="replace")
        lines = content.splitlines()[:n_lines]
        print("\n".join(lines))
    except KeyError:
        print("File not found in archive.")
```

First, we examine the original QASM generated by Qiskit’s random circuit generator:

```python
# Open the archive and display the original file
import zipfile
with zipfile.ZipFile(archive_path, "r") as zf:
    print_head_from_zip(zf, raw_path)
```

Next, we examine the transpiled version of the same circuit:

```python
# Open the archive and display the transpiled file
with zipfile.ZipFile(archive_path, "r") as zf:
    print_head_from_zip(zf, transpiled_path)
```

# Accessing QASM Code for Real Quantum Device

This notebook also explores the contents of the archive `../data/qasm_real_archive.zip`, which contains circuits used in real-device experiments.

```python
archive_path = Path("../data/qasm_real_archive.zip")

# specify the parameters needed to access the QASM listing
n_qubits = 40
algorithm = "ghz" # or "wstate"

# Construct filename
file_path = f"{algorithm}_alg_{n_qubits}.qasm"

print("Real circuit file path: ", file_path)
```

```python
# Open archive and display both files
with zipfile.ZipFile(archive_path, "r") as zf:
    print_head_from_zip(zf, file_path)
```
